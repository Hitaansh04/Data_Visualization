# b) Discrete Variables

```python
discrete_var=['country','sex','age','generation']
# defining entropy function
def pandas_entropy(column):
  vc = pd.Series(column).value_counts(normalize=True, sort=False)
  # exponent as base
  base = e
  # applying formula for entropy
  return -(vc * np.log(vc)/np.log(base)).sum()
for col in discrete_var:  
  # printing mode using df.mode() and entropy with made function
    print("Mode of %s : %s" % (str(col),str(df[col].mode().values[0])))
    print('Entropy : ', pandas_entropy(df[col]))
    print()
```

# Q 2(a)

```python
Countries=df.country.unique()
display(Countries)
```

```python
# Function to get average yearly suicides for a specific country
def get_average(df,var):
  get_suicides=df.loc[(df['country'])==var]
  
  total_suicides=get_suicides.groupby(df['year'],as_index=False)
  sum=0.0
  years=1
  for key, item in total_suicides:
    years=years+1
    # summing over suicides for a country in a year across all categories
    total=total_suicides.get_group(key)['suicides_no'].sum()
    sum+=total
  # average  
  sum=sum/years
  return sum
  
```

```python
temp=get_average(df,'Albania')
print(temp)
```

```python
# initialize dictionary
d={}
for i in Countries:
  d[i]=None
# key in the dictionary represents countries and the value associated is the average suicides
for i in d:
  average=get_average(df,i)
  d[i]=average
```

```python
import operator
# sorting in descending order by value
sorted_tuples = sorted(d.items(), key=operator.itemgetter(1),reverse=True)
print(sorted_tuples)
```

```python
highest=[]
values=[]
# taking top 6 siz countries by suicides and storing them in a seperate list
for i in range(0,6):
  highest.append(sorted_tuples[i][0])
  values.append(sorted_tuples[i][1])
print(values)
```

```python
# function for plotting the average suicide rates of top 6 
def plothist(x,y):
    for i in range(len(x)):
        # text at the top of the histogram specifying the exact value
        plt.text(i,y[i].round(),y[i].round(),ha='center',Bbox = dict(facecolor = 'orange', alpha =0.8))
plt.figure(figsize=(12,12))
plt.bar(highest,values,color=[(0.3,0.4,0.5),'Blue','Red','Yellow','Black','cyan'])
plt.title("Suicide Rates ",fontsize=20)
plt.xlabel("Countries",fontsize=16)
plt.ylabel("Average Yearly Suicides",fontsize=16)
plothist(highest,values)
plt.show()
```

```python
# function for plotting suicides of a country for each year seperately
def country_plots(df,var):
  get_suicides=df.loc[(df['country'])==var]
  total_suicides=get_suicides.groupby(df['year'],as_index=False)
  
  year=[]
  suicides=[]
  for key, item in total_suicides:
    
    total=total_suicides.get_group(key)['suicides_no'].sum()
    year.append(key)
    suicides.append(total)
  
  f=plt.figure(figsize=(12,12))
  
  ax = sns.barplot(year,y=suicides)
  ax.set(xlabel='')
  plt.title(var,weight='bold')
  #adding a common x and y label
  f.text(0.52,-0.01, "Year", ha="center", va="center",size='xx-large')
  f.text(-0.01,0.50, "Suicides", ha="center", va="center", rotation=90,size='xx-large')
  plt.tight_layout()
  plt.show()
  
```

```python
for i in highest:
  country_plots(df,i)
```

Except United States, all countries have had their suicides peaking in the late 90's or early 2000's and gradually decreasing going forward.

```python
US_df=df.loc[df['country']=='United States']
# Grouping the total number of suicides per year by generation for the US
US_gen = US_df.groupby(['year', 'generation'])['suicides_no'].sum().reset_index(name ='generational_suicides')
gen=pd.unique(US_gen['generation'])

for col in gen:
  # Getting records specific to each generation
  df_latest=US_gen.loc[US_gen['generation']==col]

  # Getting the maximum value of the number of suicides
  maximum=df_latest['generational_suicides'].max()
  # Getting the record corresponding to the maximum value of the number of suicides
  maxYear=df_latest.loc[df_latest['generational_suicides'] == maximum]
  
  print("The worst year for %s is : %s" % (col,maxYear['year']))
  plt.figure(figsize = (13, 8))
  # Plotting a line plot
  sns.lineplot(x = df_latest['year'], y = df_latest['generational_suicides'],color='maroon', lw=3)
  # Setting various axis parameters
  plt.tick_params(axis='x', which='major', labelsize=12)
    
  plt.xlabel("Year", fontsize = 14)
    
  plt.tick_params(axis='y', which='major', labelsize=12)
    
  plt.ylabel("Number of Suicides ({})".format(col), fontsize = 14)

  # Plotting a point to highlight the maximum number of suicides for the particular generation
  plt.plot(maxYear['year'], maxYear['generational_suicides'], 'ro')

  # Drawing a vertical line through highighted point for better visibility
  plt.axvline(x = int(maxYear['year']), color='k', linestyle='-')
    
  plt.show()
```

```python
df2 = df.groupby(['year','country','sex','age']).sum()
# finding an index from multiindex using df.xs
df3 = df2.xs(2005,level='year')
# converting multiindex to normal index
df3 = df3.reset_index()
# selecting some countries for visualization
countries = ['Albania','Argentina','Japan','Uzbekistan','Thailand' ]
for country in countries:
    #selecting countries one by one
    df4 = df3.loc[df3['country']==country]
    #Selecting population and age columns of male and female data
    df_male = df4.loc[df4['sex']=='male'][['population','age']]
    df_female = df4.loc[df4['sex']=='female'][['population','age']]
    # merging male female data by age data to plot age-wise
    df_merge = pd.merge(df_male, df_female, on='age')
    df_male.reset_index(drop=True,inplace=True)
    df_female.reset_index(drop=True,inplace=True)
    df_merge = df_merge[['age','population_x','population_y']]
    # For bihist plot creating subplot with 2 rows and 1 column
    fig, axes = plt.subplots(nrows=2, ncols=1, sharex=True, figsize=(10,5))
    # selecting an axis, color and data for each plot
    for ax, column, color, invert in zip(axes.ravel(), ['population_x','population_y'], ['blue', 'magenta'],  [False,True]):
        ax.bar(df_merge['age'],df_merge[column],  color=color)
        plt.xticks(rotation=90)
        ax.set(xlabel='age group',ylabel='Population')
        # inverting if counter invert is true(alternative)
        if invert:
            ax.invert_yaxis()
        # alternatively plotting title as we need one title for bihist
        else :
            ax.set_title(country, fontweight='bold')
    # To stick bihists together making hsoacce between them=0
    plt.subplots_adjust(hspace=0)
```

We can see Japan when compared to others has unusually high population in 55-74 years bracket which are retired or on the verge of retiring. This can be alarming for in the coming years there will be a shortage of working class.

# Q2(F)

```python
# dropping Nan values in HDI column
df_HDI = df.dropna(subset=['HDI for year'],how='any')
# Grouping the suicides per 100k and population by the HDI values per year for each country
df_HDI_grouped = df_HDI.groupby(['country', 'year', 'HDI for year'])['suicides/100k pop', 'population'].sum()
df_HDI_grouped.reset_index(inplace=True)
#display(df_HDI_grouped)

# Averaging out the HDI values, population and suicides/100k values per year
df_avg_hdi_grouped = df_HDI_grouped.groupby(['country'])['HDI for year', 'suicides/100k pop', 'population'].mean()

df_avg_hdi_grouped.reset_index(inplace=True)

df_avg_hdi_grouped['population'] = np.log(df_avg_hdi_grouped['population'])
display(df_avg_hdi_grouped)
```

```python
# Setting the figure size
plt.figure(figsize = (15, 8))

# Plotting a bubble plot of HDI per year vs suicides/100k with population deciding the size of the bubble
bubble = sns.scatterplot(data = df_avg_hdi_grouped, x = 'HDI for year', y = 'suicides/100k pop', hue = 'population', size = 'population', sizes=(200, 1000))

# Positioning the legend
bubble.legend(loc='center left', bbox_to_anchor=(0.8, 0.8), ncol=1)

# Setting various axis parameters
plt.tick_params(axis='x', which='major', labelsize=12)

# Setting various axis parameters   
plt.xlabel("Average HDI Per Year", fontsize = 14)
    
plt.tick_params(axis='y', which='major', labelsize=12)
    
plt.ylabel("Average Suicides/100k Per Year", fontsize = 14)

plt.show()
```

An interesting observation that can be drawn from this bubble plot is that as the average HDI index increases, the average suicides/100k population also inceases in general. This is opposite to our intuition because, an increased HDI signals more prosperity with better living conditions for all people, and yet the number of suicides for such countries is higher.
